## Plotting using matplotlib

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4*np.pi, 64)
plt.plot(x, np.sin(x), color="blue", linewidth=2, linestyle="-")
plt.plot(x, np.cos(x), color="red", linewidth= 3, linestyle="-")
plt.show()
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

N = 16
X, Y = np.meshgrid(np.arange(N), np.arange(N))
T = np.arctan2(Y, X)


plt.scatter(X, Y, s=75, c=T, alpha= 1, edgecolors='black', marker='o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 16
X, Y = np.meshgrid(np.arange(N), np.arange(N))

u_x = .5 * Y
u_y = 0
X = X + u_x
Y = Y + u_y

T = np.arctan2(Y, X)


plt.scatter(X, Y, s=75, c=T, alpha= 1, edgecolors='black', marker='o')
plt.show()
```

# Subplot in the 3d plots using matplotlib.

```python
%matplotlib inline
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

xx = np.linspace(-8, 8, 256)
yy = xx
x, y = np.meshgrid(xx, yy)

f = plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

# plot 1
ax = f.add_subplot(2, 2, 1, projection='3d', )
r = np.sqrt(x**2 + y**2)
c = 0
r1 = abs(r + c)
r2 = abs(r - c)
heights = 0.5*(r1*np.exp(-r1*r1) + r2*np.exp(-r2*r2))
ax.plot_surface(x, y, heights, rstride=1, cstride=1, cmap=cm.coolwarm,
        linewidth=0, antialiased=False)
#plot 2
ax = f.add_subplot(2, 2, 2, projection='3d')
c = 1
r1 = abs(r + c)
r2 = abs(r - c)
heights = 0.5*(r1*np.exp(-r1*r1) + r2*np.exp(-r2*r2))
ax.plot_surface(x, y, heights, rstride=1, cstride=1, cmap=cm.coolwarm,
        linewidth=0, antialiased=False)
#plot 3
ax = f.add_subplot(2, 2, 3, projection='3d')
c = 4
r1 = abs(r + c)
r2 = abs(r - c)
heights = 0.5*(r1*np.exp(-r1*r1) + r2*np.exp(-r2*r2))
ax.plot_surface(x, y, heights, rstride=1, cstride=1, cmap=cm.coolwarm,
        linewidth=0, antialiased=False)
#plot 4
ax = f.add_subplot(2, 2, 4, projection='3d')
c = 8
r1 = abs(r + c)
r2 = abs(r - c)
heights = 0.5*(r1*np.exp(-r1*r1) + r2*np.exp(-r2*r2))
ax.plot_surface(x, y, heights, rstride=1, cstride=1, cmap=cm.coolwarm,
        linewidth=0, antialiased=False)
```
