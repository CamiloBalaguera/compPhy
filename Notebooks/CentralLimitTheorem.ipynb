## Gaussian distribution
$P(x;\mu,\sigma)=\displaystyle \frac{1}{\sqrt{2 \pi \sigma^2}} \exp{\displaystyle \left( -\frac{(x-\mu)^2}{2 \sigma^2} \right) },
\hspace{1in} x \in [-\infty;\infty]$

### mean=$\mu$, variance=$\sigma^2$
### Central Limit Theorem

```python
%matplotlib inline
u=5 # mean
s=1 # standard deviation
x=np.linspace(0,10, 64)

y=(1/(np.sqrt(2*np.pi*s*s)))*np.exp(-(((x-u)**2)/(2*s*s)))

plt.plot(x,y,'o-')
plt.title('Gaussian: $\mu$=%.1f, $\sigma$=%.1f' % (u,s),fontsize=15)
plt.xlabel('x',fontsize=15)
plt.ylabel('Probability density',fontsize=15)
plt.show()
```

## Binomial distribution
$$
P(n;N,p)=\displaystyle \left(
\frac{N!}{n!(N-n)!}
\right)
p^n (1-p)^{N-n}
$$

### mean=$Np$, variance=$Np(1-p)$


```python
N = 60
p = 0.5
n = np.arange(0,200)

y = stats.binom.pmf(n,N,p)


plt.plot(n,y,'o-')
plt.title('Binomial: N=%i , p=%.2f' % (N,p),fontsize=15)
plt.xlabel('n',fontsize=15)
plt.ylabel('Probability of n',fontsize=15)
plt.show()
```

## Poisson distribution
$$
P(n;\mu)=\frac{\mu^n e^{-\mu}}{n!}
$$

### mean=$\mu$, variance=$\mu$


```python
u=600
n=range(0,1000)

y=stats.poisson.pmf(n,u)

plt.plot(n,y,'o-')
plt.title('Poisson: $\mu$ =%i' % u,fontsize=15)
plt.xlabel('n',fontsize=15)
plt.ylabel('Probability of n',fontsize=15)
plt.show()
```

## Exponential

$P(x;k)=k e^{-kx}, \hspace{1in} x \in [0,\infty], \hspace{0.1in} k>0$

### mean=$1/k$,  variance=$1/k^2$

```python
k=0.4
x=np.linspace(0,15,64)
y=k*np.exp(-k*x)
plt.plot(x,y,'o-')

plt.title('Exponential: $k$ =%.2f' % k,fontsize=15)

plt.xlabel('x',fontsize=15)
plt.ylabel('Probability density',fontsize=15)
plt.show()
```

## Beta
 $P(x;\alpha,\beta)=\displaystyle \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha) \Gamma(\beta)}x^{\alpha-1}(1-x)^{\beta-1}$, $\hspace{1in} x \in  [0;1], \hspace{0.1in} \alpha>0, \hspace{0.1in}\beta>0$

* mean=$\displaystyle \frac{\alpha}{\alpha+\beta}$, variance=$\displaystyle \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta +1)}$


```python
a=0.5
b=0.5
x=np.linspace(0.01,1,64)

y=stats.beta.pdf(x,a,b)

plt.plot(x,y,'-')
plt.title('Beta: a=%.1f, b=%.1f' % (a,b) , fontsize=15)
plt.xlabel('x',fontsize=15)
plt.ylabel('Probability density',fontsize=15)
plt.show()
```

## Central limit theorem

Let  $\{X_n\}$ be a sequence of random variables. A sequence of independent and identically distributed (iid) random variables.
> Central Limit Theorem (CLT) says that the sum of iid random variables approaches a Gaussian distribution, as the sample size increases.
$$ $$
As $n$ gets bigger the distribution of the sample mean will always converge to a Gaussian (normal) distribution with mean $\mu$ and standard deviation $\sigma/\sqrt{n}$.



```python
%matplotlib inline
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
nr = 1 # number of random variables added together 
N = 10000000  # number of samples taken
x = np.zeros((N))
for i in range(N):    
    for j in range(nr):
        x[i] += np.random.random() 
    x[i] *= 1/nr
plt.hist(x, 100, normed=True);
binrange = np.linspace(0, 1, 64)
x = np.linspace(0, 1, 128)
mean = 0.5
variance = (1/(12*nr))# \sigma^2 = \sigma^2/nr
fac = 1/np.sqrt(2*np.pi*variance)
dist = fac*np.exp(-(x-mean)**2/(2*variance))
plt.plot(x, dist, 'r-')
# the red line is the behaviour, as expected by the central limit theorem
```
