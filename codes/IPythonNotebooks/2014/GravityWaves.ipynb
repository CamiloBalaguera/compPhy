## Motion of the fluid drop as a function of wavelength
### $$\lambda = 0.1 h$$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
f = plt.figure(num=None, figsize=(15, 10), dpi=80, facecolor='w', edgecolor='k')

A = 1.5
h = 2700
z1 = -0
waveLength = 0.1*h
k = 2*pi/waveLength
x1 = linspace(0, waveLength, 64)

ax = f.add_subplot(2, 2, 2)
x = A*cosh(k*(h+z1))*cos(k*x1)
z = A*sinh(k*(h+z1))*sin(k*x1)
ax.plot(x,z)
axis('tight')
#axis([-2, 2, -2, 2])
#axis('off')
savefig('wlength=1e-1h_z1=0.png')
```

### $$\lambda =  h$$
The wavelength is depth. The motion starts is circular throughout.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
f = plt.figure(num=None, figsize=(15, 15), dpi=80, facecolor='w', edgecolor='k')

A = 2
h = 2700

waveLength = 1*h
k = 2*pi/waveLength
x1 = linspace(0, waveLength, 64)

z1 = -0
ax = f.add_subplot(2, 2, 1)
x = A*cosh(k*(h+z1))*cos(k*x1)
z = A*sinh(k*(h+z1))*sin(k*x1)
ax.plot(x,z)
axis('tight')
axis([-600, 600, -600, 600])
#axis('off')
title('Depth = 0')


z1 = -h/10
ax = f.add_subplot(2, 2, 2)
x = A*cosh(k*(h+z1))*cos(k*x1)
z = A*sinh(k*(h+z1))*sin(k*x1)
ax.plot(x,z)
axis('tight')
axis([-600, 600, -600, 600])
title('Depth = h/10')

z1 = -h/4
ax = f.add_subplot(2, 2, 3)
x = A*cosh(k*(h+z1))*cos(k*x1)
z = A*sinh(k*(h+z1))*sin(k*x1)
ax.plot(x,z)
axis('tight')
axis([-600, 600, -600, 600])
title('Depth = h/4')

z1 = -h*0.6
ax = f.add_subplot(2, 2, 4)
x = A*cosh(k*(h+z1))*cos(k*x1)
z = A*sinh(k*(h+z1))*sin(k*x1)
ax.plot(x,z)
axis('tight')
axis([-600, 600, -600, 600])
title('Depth = 0.6*h')


savefig('wlength=h.png')
```

### $$\lambda = 10 h$$
Here the whole body of the fluids from z=0 to z=-h is in motion as the wavelength is 10 times the depth. The motion starts becoming more and more elliptical as u go along the bottom.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
f = plt.figure(num=None, figsize=(15, 15), dpi=80, facecolor='w', edgecolor='k')

A = 1.5
h = 2700

waveLength = 10*h
k = 2*pi/waveLength
x1 = linspace(0, waveLength, 64)

z1 = -0
ax = f.add_subplot(2, 2, 1)
x = A*cosh(k*(h+z1))*cos(k*x1)
z = A*sinh(k*(h+z1))*sin(k*x1)
ax.plot(x,z)
axis('tight')
axis([-2, 2, -2, 2])
#axis('off')
title('Depth = 0')


z1 = -h/2
ax = f.add_subplot(2, 2, 2)
x = A*cosh(k*(h+z1))*cos(k*x1)
z = A*sinh(k*(h+z1))*sin(k*x1)
ax.plot(x,z)
axis('tight')
axis([-2, 2, -2, 2])
#axis('off')
title('Depth = -h/2')

z1 = -0.9*h
ax = f.add_subplot(2, 2, 3)
x = A*cosh(k*(h+z1))*cos(k*x1)
z = A*sinh(k*(h+z1))*sin(k*x1)
ax.plot(x,z)
axis('tight')
axis([-2, 2, -2, 2])
title('Depth = 0.9*h')

#axis('off')


z1 = -h
ax = f.add_subplot(2, 2, 4)
x = A*cosh(k*(h+z1))*cos(k*x1)
z = A*sinh(k*(h+z1))*sin(k*x1)
ax.plot(x,z)
axis('tight')
axis([-2, 2, -2, 2])
#axis('off')
title('Depth = h')


savefig('wlength=10h.png')
```
