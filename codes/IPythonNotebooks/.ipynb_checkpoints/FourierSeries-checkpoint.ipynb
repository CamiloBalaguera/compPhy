## Fourier methods

Fourier transform (FT) for a well-behaved functions $f$ is defined as:

$$f(k) = \int e^{ikx} f(x) ~dx$$

The inverse FT is then 

$$f(x) = \frac{1}{2\pi} \int e^{ikx} f(k) ~dk$$



## Discrete fourier tranforms (DFTs)


If the function is peridoc in real space, $f(x+L) = f(x)$, then the Fourier space is discrete. 

$$f(x) = \sum e^{ikx} f(k) ~dk~~~~~~$      where $k = 0,~ \pm\frac{2\pi}{L},~\pm2\frac{2\pi}{L}, ... $$


Usually in computations we need to find FT of discrete function rather than of a well defined analytic function. So suppose if the function is defined at $N$ points in ral space and u want to calculate the function at $N$ points in Fourier space. The DFT is defined as 

$$f_k = \sum_{n=0}^{N-1} f_n ~ e^{i~2\pi~k~n~/~N}$$

while the inverse of this called the 

$$f_n = \sum_{n=0}^{N-1} f_k ~ e^{i~2\pi~k~n~/~N}$$

To calculate each $f_n$ one needs $N$ computations and it has to be done $N$ times, i.e, the algorithm is simply $\mathcal{O}(N^2)$.


## Fast fourier tranforms (FFTs)
