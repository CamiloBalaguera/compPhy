## Fourier methods

Fourier transform (FT) for a well-behaved functions $f$ is defined as:

$$f(k) = \int e^{ikx} f(x) ~dx$$

The inverse FT is then 

$$f(x) = \frac{1}{2\pi} \int e^{ikx} f(k) ~dk$$



## Discrete fourier tranforms (DFTs)


If the function is periodic in real space, $f(x+L) = f(x)$, as well discrete with the spacing $h$, then the Fourier space is discrete with spacing $\frac{2\pi}{L}$. 

$$f(x) = \sum e^{ikx} f(k) ~dk~~~~~~      \text{where } k = \Bigg[ -\frac{\pi}{h}, \frac{\pi}{h}\Bigg];~~ \text{with interval} \frac{2\pi}{L} $$


So we see that there is a concept of the maximum wavenumber $ k_{max}=\frac{\pi}{h} $, we will get back to this later in the notes. Usually in computations we need to find FT of discrete function rather than of a well defined analytic function. Since the real space is periodic and discrete, the Fourier space is both bounded and discrete. Also, the Fourier space is continous if the real space is unbounded. If the function is defined at $N$ points in real space and one wants to calculate the function at $N$ points in Fourier space, then DFT is defined as 

$$f_k = \sum_{n=0}^{N-1} f_n ~ e^{i~2\pi~k~n~/~N}$$

while the inverse of this is

$$f_n = \sum_{n=0}^{N-1} f_k ~ e^{i~2\pi~k~n~/~N}$$

To calculate each $f_n$ one needs $N$ computations and it has to be done $N$ times, i.e, the algorithm is simply $\mathcal{O}(N^2)$. This can be implemented numerically as a matrix multiplication, $f_k = M\cdot f_n$, where $M$ is a $N\times N$ matrix.


## Fast fourier tranforms (FFTs)

FFTs improves on DFTs by exploiting their symmetries. 

$$ \begin{align}
f_k &= \sum_{n=0}^{N-1} f_n  e^{-i~\frac{2\pi~k~n}{N}} \\
&= \sum_{n=0}^{N/2-1} f_{2n} e^{-i~\frac{2\pi~k~2n}{N}} &+ \sum_{n=0}^{N/2-1} f_{2n + 1} e^{-i~\frac{2\pi~k~(n+1)}{N}}\\
&= \sum_{n=0}^{N/2 - 1} f_{2n} e^{-i~\frac{2\pi k~n}{N/2}} &+ e^{-i\frac{2\pi}{N}} \sum_{n=0}^{N/2 - 1} f_{2n + 1} e^{-i~\frac{2\pi~k~n~}{N/2}}\\
&=\vdots &\vdots
\end{align}$$

Because of the tree structure the complexity of the problem becomes $\mathcal{O}(N \ln N)$ which becomes a major advantage over the $\mathcal{O}(N^2)$ algorithm. Below we will look at a simple implementation of FFT

```python
from __future__ import division
import numpy as np
```

```python
def testFFT(x):
    ''' FFT in 1d  '''
    N = x.shape[0]
    if N % 2 > 0:
        raise Exception('x must have even size')
    elif N <= 16: 
        ''' this is the naive implementation using matrix multiplication'''
        n = np.arange(N)
        k = n.reshape((N, 1))
        M = np.exp(-2j * np.pi * k * n / N)
        return np.dot(M, x)
    else:
        M1 = testFFT(x[::2])
        M2 = testFFT(x[1::2])
        fac = np.exp(-2j * np.pi * np.arange(N) / N)
        return np.concatenate([M1+fac[:N/2] * M2, M1+fac[N/2:]*M2])
```

```python
x = np.random.random(1024)
np.allclose(np.fft.fftn(x), testFFT(x))
```

## More examples
### Sampling: Aliasing error
There is this error becuase of the $k_{max}$ since if the spacing is $h$ then $k>k_{max}$ can not be distinguished.
In the example below

```python
L, N = 1, 9
x = linspace(-L, L, 512)
xx = linspace(-L, L, N)

def f1(x):
    return sin((pi/L)*x)

def f2(x):
    return sin((17*pi/L)*x)
```

```python
f = plt.figure(figsize=(17, 6), dpi=80);  

plot(x, f1(x), color="#A60628", linewidth=2);
plot(x, f2(x), color="#348ABD", linewidth=2);
plot(xx, f1(xx), 'o', color="#020e3e", markersize=8)
xlabel('x', fontsize=15); ylabel('y(x)', fontsize=15);
title('Aliasing in sampling of $sin(\pi x/L)$ and $sin(17 \pi x/L)$', fontsize=15);
```

 ### Differentiation

```python
from __future__ import division
def f1(x):
    return cos(2*pi*x/L)
```

```python
f = plt.figure(figsize=(10, 5), dpi=80);  

L, N = 2, 32
kk   = 2*pi/L
x = arange(0, N)*(L/N)     
k = np.concatenate(( np.arange(0, N/2+1,1), np.arange(-N/2+1, 0, 1) ))*kk 

fk = np.fft.fft(f1(x))         
f1_kk = -k*k*fk               
f1_xx = np.fft.ifft(f1_kk)

 
plot(x, -f1(x)*kk*kk, color="#348ABD", label = 'analytical', linewidth=2)          
plot(x, f1_xx, 'o', color="#A60628", label = 'numerical', markersize=6)     
legend(loc = 'best')
xlabel('x', fontsize=15)
title('derivatives using FFT', fontsize=15)
xlim([0, max(x)]);
```

```python
# since we knwo that functions are real then we could save some time and memory by using real FFTs.
f = plt.figure(figsize=(10, 5), dpi=80);  

L, N = 1, 32            
x = arange(0, N)*(L/N)  
kk   = 2*pi/L
k = kk* arange(0, N/2 + 1)

fk = np.fft.rfft(f1(x))         
fk = -k*k*fk               
fx = np.fft.irfft(fk)

plot(x, -kk*kk*f1(x), color="#348ABD", label = 'analytical', linewidth=2)          
plot(x, fx, 'o', color="#A60628", label = 'numerical')     
legend(loc = 'best')
xlabel('x', fontsize=15)
title('derivatives using FFT', fontsize=15)
xlim([0, max(x)]);
```
