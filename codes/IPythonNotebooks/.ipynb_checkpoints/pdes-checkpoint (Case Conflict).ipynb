## Solving partial differential equations
Differential equations are ubiquitous in physics. Also, in many interesting cases they are not tractable analytically. So we take recourse to numerical methods, every now and then, to solve these differential equations. In this talk we will discuss some basics of solving PDEs on the computer. Some important equations are


|Equation|Mathematical form                            |Type       |
|:----------:|:-----------------------------------------:|:---------:|
|Laplace   |  $ \nabla^2 u = 0 $                       |Elliptic   |
|Poisson   |  $ \nabla^2 u = \frac{\rho}{\epsilon_0} $ |Elliptic   |
|Heat      |  $ \partial_t u = \nabla^2 u  $           |Parabolic  |
|Wave      |  $ \partial_t^2 u = \nabla^2 u  $         |Hyperbolic |
|Helmholtz |  $ (\nabla^2 + k^2) u = 0  $              |Elliptic   |
|$\vdots$  |  $\vdots  $                               |$\vdots$   |

### Forward difference
Let denote the space derivative by '. So, $\nabla^2  f(x)= f''(x) $. The first derivative of the function can be written in discrete form as

$$  f'(x)  \approx \frac{f(x+h)-f(x)}{h} $$

This is called forward difference as this uses $f(x)$ and $f(x+h)$.  

$$  f'(x)  = \frac{f(x+h)-f(x)}{h} + \mathcal{O}(h)$$

### Backward difference
$$  f'(x)  \approx \frac{f(x)-f(x-h)}{h} $$

This way to calculate first derivative is called backward difference as this uses $f(x)$ and $f(x-h)$. 

$$  f'(x)  = \frac{f(x)-f(x-h)}{h} + \mathcal{O}(h)$$

### Central difference
$$  f'(x)  \approx \frac{f(x+h)-f(x-h)}{2h} $$
The error in this case is $ \mathcal{O}(h^2)$$



```python
#parameters
a, b, h_, N = 0, 10, 1, 32

def forwardDifference(f, h=h_):
    '''     f' = ( f(x+h)-f(x) )/h + O(h)   '''
    return ( f(x+h)-f(x) )/h

def backwardDifference(f, h=h_):
    '''    f' = ( f(x)-f(x-h) )/h + O(h)    '''
    return ( f(x)-f(x-h) )/h

def centralDifference(f, h=h_):
    '''    f' = (f(x+h) - f(x-h))/h + O(h^2)'''
    return (f(x+h) - f(x-h))/(2*h)

# test Function
def f(x):
    return np.exp(x)

#Main body
x_ = linspace(a, b, N)
forward = []
forward_h2 = []
central = [];

for x in x_:
    forward.append( forwardDifference(f) )
    forward_h2.append(forwardDifference(f, h=1e-2) )
    central.append( centralDifference(f) )

# Plotting
figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
axis([a,b,0,np.exp(b)])
plot(x_,forward,'D',label='Forward h = %g'%h_)
plot(x_,forward_h2,'o', color="#A60628", label='Forward h = %g'% 1e-2)
plot(x_,central,'^', color='#348ABD', label='Central  h = %g'%h_)
plot(x_,f(x_),'-',label='Exact', color="#348ABD", linewidth=2)
 
legend(loc='upper left', fontsize=20)
xlabel("x", fontsize=20)
ylabel("$f$ '$(x)$", fontsize=20)
title("$f$ '$(x)$ under forward and central difference for $f(x)=exp(x)$", fontsize=20);
```

## Diffusion equation

Lets solve diffusion using the machinery we have developed above.

$$ \partial_t u = \nabla^2 u  $$    

The steps involved are

 * construct a discretized Laplacian
 * evolve in time

Laplacian can be constructed bu applying the central difference twice, for step size $h/2$ 

$$\nabla^2  f(x)=f''(x) \approx \frac{f(x_{i-1}) - 2 f(x_i) + f(x_{i+1})}{h^2}$$

```python
## Class to simulate the diffusion equation
from __future__ import division
class simulateDiffusion(): 
    def __init__(self):
        pass

    def Gaussian(self, mean,sigma,x):
        ''' Returns Gaussian distribution''' 
        return 1./(sigma*sqrt(2*np.pi))*exp(-0.5*(x-mean)**2/sigma**2)
    
    def Laplacian(self, u, h ):
        ''' calculates the Laplcian of a  given u(x,t)'''
        Laplacian_ = zeros(u.shape, np.float) 
        
        for i in range(1, len(u)-1):
            Laplacian_[i] = (u[i+1] - 2*u[i]+u[i-1])/h**2
        
        # at the boundaries we assume that u does not change u[0]=u[1] and u[N]= u[N-1]
        i = 0
        Laplacian_[i] = (u[i+1] - 2*u[i]+u[i])/h**2
        i = -1 # or len(u)-1
        Laplacian_[i] = (u[i] - 2*u[i]+u[i-1])/h**2
        return Laplacian_
        
   
    def simulate(self, u, x, a, D, T, h):
        ''' simulates and gives the snapshots of the configurations for some time instant '''
        figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
        for i in range(T):
            f_ = D*self.Laplacian(u, h)
            u = u + dt*f_
                    
            if i==  0  or i==int(T/8) or i==int(T/4) or i==int(T/2):# or i==T-1:
                axis([-a, a, 0, 1])  
                plot(x,u,'-o', color="#A60628", alpha= .1 + (1.5*i/T), label='Time = %g'%i, linewidth=3)   
         
        legend(loc='upper left', fontsize=20)
        xlabel("x", fontsize=20)
        ylabel("u(x)", fontsize=20)
        ylim([0, 0.6])
        title("Diffusion of a Gaussian distribution at origin at T=0.", fontsize=20)
        show()
```

```python
## This block simulates the diffusion equation for an initial Gaussian disturbance at the origin in 1d
from __future__ import division
# parameters
a, N, T, dt = 10, 128, 400, 0.01
x = linspace(-a, a, N) 
h = x[1]-x[0]       
D = 1.                         #Diffusion coefficient

# instantiate the class
rm = simulateDiffusion()

# initial configuration
u = rm.Gaussian(mean=0.,sigma=.75,x=x)

# Simulate 
rm.simulate(u, x, a, D, T, h)

## Simulation completed!
```

## Using matrices

All these can be done very conveniently using matrices. This has been done very nicely [here](https://github.com/ketch/finite-difference-course) by David Ketcheson. We will follow some of the techniques used by him.

The central difference Laplacian  

$$\nabla^2  f(x)=f''(x) \approx \frac{f(x_{i-1}) - 2 f(x_i) + f(x_{i+1})}{h^2}$$

can be very conveniently written in matrix form as

$$A= \frac{1}{h^2} \left( \begin{array} -2 & 1 \\\\\\\\ 1 & -2 & 1 \\\\\\\\ & 1 & -2 & 1 \\\\\\\\ & &  \ddots &  \ddots &  \ddots \\\\\\\\        & & & 1 & -2 & 1 \\\\\\\\ & & & & 1 & -2\end{array} \right),$$

So we could have directly created this matrix.

```python
m=5
e=np.ones(m)
print np.diag(e,0)
```

```python
e1=np.ones(m-1)
print np.diag(e,-1)
print np.diag(e,1)
```

```python
def A_matrix(m):
    e=np.ones(m)
    e1=np.ones(m-1)
    A = np.diag(e1,-1)- 2*np.diag(e,0) + np.diag(e1,1)
    return A
```

```python
A_matrix(2)
```

## 5-point Laplacian finite difference formula 
$$\frac{1}{h^2} \left(f_{i-1,j}+f_{i+1,j}+f_{i,j-1}+f_{i,j+1}-4f_{ij} \right)$$   


```python
def five_pt_laplacian(m):
    e=np.ones(m**2)
    e2=([0]+[1]*(m-1))*m
    h=1./(m+1)
    A=np.diag(-4*e,0)+np.diag(e2[1:],-1)+np.diag(e2[1:],1)+np.diag(e[m:],m)+np.diag(e[m:],-m)
    A/=h**2
    return A
```

```python
A = five_pt_laplacian(4)
spy(A)
```

## Sparse matrix
Taken from [here.](http://nbviewer.ipython.org/github/ketch/finite-difference-course/blob/master/ipython/Week_3_lab_2D_BVPs-with_solutions.ipynb) Storing a big matrix will take up a lot of memory, so we define a spdiags matrix which only stores the non-zero elements of the matrix.


```python
#Storing the above will take up a lot of memory, so we define a spdiags matrix!
from scipy.sparse import spdiags,linalg
def five_pt_laplacian_sparse(m):
    e=np.ones(m**2)
    e2=([1]*(m-1)+[0])*m
    e3=([0]+[1]*(m-1))*m
    h=1./(m+1)
    A=spdiags([-4*e,e2,e3,e,e],[0,-1,1,-m,m],m**2,m**2)
    A/=h**2
    return A
```

```python
A = five_pt_laplacian_sparse(4)
spy(A, color='#A60628', alpha=0.6)
```

## Example
Lets solve the following PDE:
  $$ u_t = k \nabla^2 u + f(u)$$
      
    
  where
  $$ f(u) = a u + b u^3 $$
  


```python
import numpy as np
from __future__ import division
from scipy.sparse import spdiags,linalg,eye
import matplotlib.pyplot as plt


a, b, k  = 0, 1.0, 100.0
dh, dt = 1.0, 1e-3
N, T  = 256, 10000

class pdeSolver():
    '''
    Class to solve a PDE 
    '''
    def mu(self, u):
        return a*u + b*u*u*u 

    def laplacian(self, N):
        """Construct a sparse matrix that applies the 5-point laplacian discretization"""
        e=np.ones(N**2)
        e2=([1]*(N-1)+[0])*N
        e3=([0]+[1]*(N-1))*N
        h=dh
        A=spdiags([-4*e,e2,e3,e,e],[0,-1,1,-N,N],N**2,N**2)
        A/=h**2
        return A

    def integrate(self, L, x, y, u):
        '''  solves the equation and plots it at differnt instants '''
        
        f = plt.figure(num=None, figsize=(15, 15), dpi=80, facecolor='w', edgecolor='k');    

        for i in range(T):          
            u = u - dt*(self.mu(u) - k*L.dot(u))
            
            if (i==0):       self.configPlot(x, y, u, f, 1, i);
            if (i==10):      self.configPlot(x, y, u, f, 2, i);
            if (i==100):     self.configPlot(x, y, u, f, 3, i);
            if (i==1000):    self.configPlot(x, y, u, f, 4, i);
            if (i==5000):    self.configPlot(x, y, u, f, 5, i);
            if (i==9999):    self.configPlot(x, y, u, f, 6, i);
     
    def configPlot(self, x, y, u,f, n_, i):
        U= u.reshape((N,N))
        sp =  f.add_subplot(3, 3, n_ )  
        plt.setp(sp.get_yticklabels(), visible=False)
        plt.setp(sp.get_xticklabels(), visible=False)
        plt.pcolormesh(x,y,U);
        plt.title('Time=%d'%i)
```

```python
rm = pdeSolver()   # instantiate the class

# generate the grid and initialise the field
x = np.linspace(-1,1,N)
y = np.linspace(-1,1,N)
X, Y = np.meshgrid(x, y)

# Initial data
u=np.random.randn(N*N, 1);

# construct the laplacian
L = rm.laplacian(N)

# Integrate
rm.integrate(L, x, y, u)

#simulation completed!!
```

## Boundary value problems (ASEP)

In this section we will discuss the asymmetric exclusion processes (ASEPs). ASEP are described by particles hooping on a lattice randomly. Here, we consider the hopping if particles in only one direction and the only once particle can sit on a lattice and hence the method is called to have exclusion. The method is asymmetric as the particles can move to different directions with distinct probabilities. In an extension to this problem, we study two parallel lanes without particle exchange. These lanes affect the dynamics indirectly by tuning the transtion rate on the respective lattices. This phenomenon is very nicely captured by the hydrodynamic description of the model. More details of the method can be found [here](http://arxiv.org/pdf/1208.4916.pdf)

In the following code we have solved the hydrodynamic equation using scikits.bvp_solver.

```python
# Solution of y1'' = f(x) and y2''(x)=g(x) over the domain (a, b)
# We have the boundary condition y1(a), y1(b), y2(a), y2(b)
#
#
# Rajesh Singh
# 20130726
#
import scikits.bvp_solver
import numpy as np
from matplotlib import pyplot as plt


def ASEP(a, b, N, b1, b2, b3, b4, epi, m, i, col):
    ''' solving the hydrodynamic approach to the ASEP problem'''
    def rhs(X, y):
        return np.array([
                         y[1],                                                               # y1'  

                         epi*( (y[1] - 2*y[0]*y[1]) * (1-m* y[2])- y[0]*(1-y[0])*m*y[3]),     # y1'' = f(x)
                         
                         y[3],                                                               # y2'
                         
                         epi*(	(y[3] - 2*y[2]*y[3]) * (1-m* y[0])- y[2]*(1-y[2])*m*y[1])    # y2'' = g(x)
                         ]) 


    def boundary_conditions(Ya, Yb):
        ''' This is to set the BC for the problem'''
        BCa = np.array([ Ya[0] - b1, Ya[2] - b3  ]) 
        
        BCb = np.array([ Yb[0] - b2, Yb[2] - b4  ]) 
        
        return BCa, BCb


    def guess_y (X):
        '''This is the guess to the answer one expects. A wise guess can help the integrator immensely!'''
        return np.array([ 
                              0.3*np.cos(X*np.pi), 
                              0.7*np.cos(X*np.pi)  , 
                              0.01*np.cos(X*np.pi), 
                              0.01*np.cos(X*np.pi)  , 
                              ])

    problem_definition = scikits.bvp_solver.ProblemDefinition(num_ODE = 4,
                                                      num_parameters = 0,
                                                      num_left_boundary_conditions = 2,
                                                      boundary_points = (a, b),
                                                      function = rhs,
                                                      boundary_conditions = boundary_conditions)

    solution = scikits.bvp_solver.solve(bvp_problem = problem_definition,
                                solution_guess = guess_y)


    x = np.linspace(a, b, N)   # define the domain of solution 
    y = solution(x)            # solve the DE and store in y
    #Simulation completed!
    
    #Plotting business
    plt.plot(x, y[0,:],'-', color=col, linewidth=3, alpha=1, label='m=%s'%(m))
    plt.legend(loc="lower left", fontsize=20)
    plt.xlabel("Lattice site (x)", fontsize=20)
    plt.ylabel("Density on the lattice site", fontsize=20)
    plt.title("Hydrodynamic approach to ASEP", fontsize=20)
##########


## define parameters
a, b, N = 0.0, 1.0, 128                         # this defines the domain 
epi   = 100                                     # singularity parameters       
b1, b2  = 0.4, 0.2                              # BC on lattice 1
b3, b4  = 0.01, 0.8                             # BC on lattice 2
m1 = np.array([0.0, 0.65,  0.725, 0.8, 1])


f = plt.figure(num=None, figsize=(18, 10), dpi=80, facecolor='w', edgecolor='k')
col = np.array(["blue", "black", "red", "#348ABD", "#A60628"])
i = 0
for i in range(5):
    ASEP(a, b, N, b1, b2, b3, b4, epi, m1[i], i, col[i])

plt.show()
```

Thus we see that the interaction between the two lattices lead to a "traffic jam" kind of stiuation om the lattice. This phenomenon has also been captured in the Monte carlo simulations.
