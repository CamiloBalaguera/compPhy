### Plotting deformation of a two dimensional grid 
Here we demonstrate that the density only changes if the divergence of the strain tensor $u$ is non-zero. 
$$
$$
$$ \nabla \cdot u \neq 0$$
$$
$$
This is essentially because of the fact that the change in volume is related to the divergence of the strain.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

N = 16
X, Y = np.meshgrid(np.arange(N), np.arange(N))
T = np.arctan2(Y, X)


plt.scatter(X, Y, s=75, c=T, alpha= 1, edgecolors='black', marker='o')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

N = 16

X, Y = np.meshgrid(np.arange(N), np.arange(N))
u_x = Y
u_y = 0
X = X + u_x
Y = Y + u_y
T = np.arctan2(Y, X)
ax1.scatter(X, Y, s=75, c=T, alpha= 1, edgecolors='black', marker='o')


X, Y = np.meshgrid(np.arange(N), np.arange(N))
u_x = X
u_y = 0
X = X + u_x
Y = Y + u_y
T = np.arctan2(Y, X)
ax2.scatter(X, Y, s=75, c=T, alpha= 1, edgecolors='black', marker='o')

```

Here we show that if the deformation is non-trivial then the body can take interesting shapes. Also, note that if the divergence of $u$ is zero then the body has only shear deformation but the density in the bulk remains constant. It is only in the case when the divergenc is non-zero we see the effect on this deformation on the bulk of the body.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

N = 20

X, Y = np.meshgrid(np.arange(N), np.arange(N))
u_x = np.cos(X)
u_y = 0
X = X + u_x
Y = Y + u_y
T = np.arctan2(Y, X)
ax1.scatter(X, Y, s=75, c=T, alpha= 1, edgecolors='black', marker='o')


X, Y = np.meshgrid(np.arange(N), np.arange(N))
u_x = np.sin(Y)
u_y = 0
X = X + u_x
Y = Y + u_y
T = np.arctan2(Y, X)
ax2.scatter(X, Y, s=75, c=T, alpha= 1, edgecolors='black', marker='o')
```

### Random numbers

### Subplot in the 3d plots using matplotlib.
    Here we give a demo of subplots in 3d using matplolib. The disturbances are are due to two different initial disturbances.

```python
%matplotlib inline
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

xx = np.linspace(-4, 4, 80)
yy = xx
x, y = np.meshgrid(xx, yy)

f = plt.figure(num=None, figsize=(15, 10), dpi=80, facecolor='w', edgecolor='k')

def plot_3dG(c_, n_):
    ax = f.add_subplot(2, 2, n_, projection='3d', )
    r = np.sqrt(x**2 + y**2)
    c = c_
    r1 = abs(r + c)
    r2 = abs(r - c)
    heights = 0.5*(np.exp(-r1*r1) + np.exp(-r2*r2))
    ax.plot_surface(x, y, heights, rstride=1, cstride=1, cmap=cm.coolwarm,
            linewidth=0, antialiased=False)

#plot 1
c, n = 0, 1
plot_3dG(c, n)

#plot 2
c, n = 2, 2
plot_3dG(c, n)

##plot 3
#c, n = 4, 3
#plot_3dG(c, n)

##plot 4
#c, n = 6, 4
#plot_3dG(c, n)

plt.show()
```
